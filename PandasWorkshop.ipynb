# Indexing

```python
#Reset the values of our index on our cleaned dataset
WBGc = WBGi.reset_index() 
WBGc
```

What are those?!?!

```python
#Great there are also periods that should actually be nulls. 
#Come on World Bank, who does that? 
#Solution: use replace and Repeat steps from above 
WBGc2 = WBGc.replace({'..': np.NaN}).dropna().reset_index() #drop option, inplace modifies the db
WBGc2
```

```python


#Index our data using the country as first heirarchy then data series
idxWBG = WBGc2.set_index(['Country Name',"Series Name"])
idxWBG
```
